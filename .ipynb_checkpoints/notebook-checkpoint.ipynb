customer order merge

```python
customers_orders_df = pd.merge(
    left=customers_df,
    right=orders_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
customers_orders_df.head()
```

#### Order_items + Seller merge

```python
item_seller_df = pd.merge(
    left=order_items_df,
    right=sellers_df,
    how="left",
    left_on="seller_id",
    right_on="seller_id"
)
item_seller_df.head()
```

#### Produt + Category merge

```python
all_product_df = pd.merge(
    left=products_df,
    right=product_category_name_translation_df,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)
all_product_df.head()
```

#### order + payment merge

```python
order_and_payment_df = pd.merge(
    left=orders_df,
    right=order_payments_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
order_and_payment_df.head()
```

#### order and payment + review merge

```python
order_payment_review_df = pd.merge(
    left=order_and_payment_df,
    right=order_reviews_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
order_payment_review_df.head()
```

#### order and payment and review + customer merge

```python
all_costumer_order_df = pd.merge(
    left=order_payment_review_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
all_costumer_order_df.head()
```

#### product and category + seller and item merge

```python
product_seler_df = pd.merge(
    left=item_seller_df,
    right=all_product_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
product_seler_df.head()
```

#### ALL merge

```python
all_df = pd.merge(
    left=all_costumer_order_df,
    right=product_seler_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
all_df.head()
```

## Visualization & Explanatory Analysis

#### 1. category dari produk apa yang paling laris dan paling sedikit diminati pelanggan? dan category manakah yang paling menguntungkan?

```python
all_df.groupby(by="product_category_name").agg({
    "order_id": "nunique",
    "payment_value": "sum"
}).sort_values(by="order_id", ascending=False)
```

```python
count_order_id_df = all_df.groupby("product_category_name")["order_id"].count().sort_values(ascending=False).reset_index()
count_order_id_df.columns = ["product_category_name", "order_id_count"]

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="order_id_count", y="product_category_name", data=count_order_id_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Performing category Product", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)

sns.barplot(x="order_id_count", y="product_category_name", data=count_order_id_df.sort_values(by="order_id_count", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Performing category Product", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Best and Worst category Product by the count of order_id", fontsize=20)

plt.show()
```

```python
sum_order_items_df = all_df.groupby("product_category_name").payment_value.sum().sort_values(ascending=False).reset_index()
sum_order_items_df.head(15)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))
 
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(x="payment_value", y="product_category_name", data=sum_order_items_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Performing category Product", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)
 
sns.barplot(x="payment_value", y="product_category_name", data=sum_order_items_df.sort_values(by="payment_value", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Performing category Product", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)
 
plt.suptitle("Best and Worst category Product by the profit", fontsize=20)
plt.show()
```

#### 2. berapa banyak pengeluaran yang dikeluarkan oleh customer dalam rentang waktu tertentu?

```python
monthly_orders_df = all_df.set_index('order_purchase_timestamp').resample(rule='M').agg({
    "customer_unique_id": "nunique",
    "payment_value": "sum"
})
start_year = 2017
end_year = 2017
filtered_orders_df = monthly_orders_df[
    (monthly_orders_df.index.year >= start_year) & 
    (monthly_orders_df.index.year <= end_year)
]
filtered_orders_df.index = filtered_orders_df.index.strftime('%B')
filtered_orders_df = filtered_orders_df.reset_index()
plt.figure(figsize=(10, 5))
plt.plot(
    filtered_orders_df["order_purchase_timestamp"],
    filtered_orders_df["payment_value"],
    marker='o', 
    linewidth=2,
    color="#72BCD4"
)
plt.title("Total pengeluaran costumer per Month (2017)", loc="center", fontsize=20)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
